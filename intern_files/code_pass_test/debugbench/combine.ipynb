```python
import pandas as pd
import os
os.chdir('/home/hadoop-aipnlp/dolphinfs_hdd_hadoop-aipnlp/liujincheng06/utils/debugbench')
df1 = pd.read_excel("mbpp_cpp_v4_with_error_info.xlsx")
```

```python
import json
code_info = []
with open('./mbpp_cpp_tests.jsonl', 'r') as f:
    for line in f.readlines():
        code_info.append(json.loads(line))
```

```python
un_passed_case = [0, 1, 2, 4, 5, 6, 8, 9, 12, 13, 14, 17, 19, 23, 24, 26, 28, 30, 31, 32, 33, 34, 35, 36, 37, 38, 40, 41, 42, 43, 44, 46, 48, 49, 51, 53, 54, 56, 57, 58, 59, 60, 61, 62, 65, 68, 70, 71, 73, 74, 75, 77, 78, 79, 80, 81, 82, 83, 84, 85, 86, 87, 88, 90, 91, 92, 94, 95]
```

```python
orig_index = []
for i in range(len(df1)):
    if df1['is_passed'][i] == 0:
        orig_index.append(i)
```

```python
len(orig_index)
```

```python
len(code_info)
```

```python
for i,v in enumerate(code_info):
    if i not in un_passed_case:
        df1['is_passed'][orig_index[i]] = 1
        df1['cpp_code_str'][orig_index[i]] = v['cpp_code_str']
        df1['cpp_test'][orig_index[i]] = v['cpp_test']
        df1['unpassed_error_info'][orig_index[i]] = ""
```
