```python
from typing import Generator
from tree_sitter import Language, Parser, Tree, Node
import os
print(os.chdir('/mnt/dolphinfs/hdd_pool/docker/user/hadoop-aipnlp/liujincheng06/task1'))
LANGUAGE = Language('my-languages.so','python')
```

```python
parser = Parser()
parser.set_language(LANGUAGE)
```

```python
def traverse_tree(tree: Tree) -> Generator[Node, None, None]:
    cursor = tree.walk()

    visited_children = False
    while True:
        if not visited_children:
            yield cursor.node
            if not cursor.goto_first_child():
                visited_children = True
        elif cursor.goto_next_sibling():
            visited_children = False
        elif not cursor.goto_parent():
            break
```

```python
import pandas as pd
import random
random.seed(42)
def replace_node_of_code(code:str,node:Node,replaced:str,diff:int,lang:str):
        code = code[:node.start_byte + diff] + replaced + code[node.end_byte + diff:]
        diff += -(node.end_byte - node.start_byte) + len(replaced)
        return code[:],diff
```

```python
df = pd.read_excel("mbpp_test.xlsx")
replaced_codes = []
replaced_types = []
for i in range(len(df)):
    code = df['code'].iloc[i]
    tree = parser.parse(bytes(code, 'utf8'))
    tree.root_node
    condidate_stmt_node = []
    condidate_identifier_node = []
    flag = False
    for node in traverse_tree(tree):
        if node.type.count("expression") > 0:
            condidate_stmt_node.append(node)
        if node.type.count("identifier") > 0 and flag:
            condidate_identifier_node.append(node)
        if node.type.count("block") > 0:
            flag = True

    len_stmt = len(condidate_stmt_node)
    len_identifer = len(condidate_identifier_node)
    diff = 0
    code_lines = len(code.split("\n"))
    length = len(replaced_types)
    if code_lines > 5:
        try:
            index1 = random.randint(0,len_stmt-1)
            code,diff = replace_node_of_code(code,condidate_stmt_node[index1],"<deleted_stmt>",diff,"python")
            replaced_types.append("delete statement")
        except:
            pass
    else:
        try:
            index2 = random.randint(0,len_identifer-1)
            code,diff = replace_node_of_code(code,condidate_identifier_node[index2],"<deleted_identifier>",diff,"python")
            replaced_types.append("delete identifier")
        except:
            pass
    if length == len(replaced_types):
        replaced_types.append("No change")
    replaced_codes.append(code)
```

```python
df['deleted_code_str'] = replaced_codes
df['delete type'] = replaced_types
```
